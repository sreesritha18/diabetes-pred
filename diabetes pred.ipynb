```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
diabetes=pd.read_csv('diabetes.csv')
diabetes.head()
```

```python
diabetes.shape
```

```python
diabetes.info()
```

```python
diabetes['Outcome'].value_counts()
```

```python
X=diabetes.drop('Outcome',axis=1)
y=diabetes['Outcome']
```

```python
X.shape
```

```python
diabetes.describe()
```

```python
scaler = StandardScaler()
standardized_data = scaler.fit_transform(X)
X = standardized_data
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.32, random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
svc=svm.SVC(kernel='linear')
svc.fit(X_train,y_train)
```

```python
train_y_pred=svc.predict(X_train)
test_y_pred=svc.predict(X_test)
```

```python
print("Train set accuracy:",accuracy_score(y_train,train_y_pred))
print("Test set accuracy:",accuracy_score(y_test,test_y_pred))
```

```python

input_data = (1, 85, 66, 29, 0, 26.6, 0.351, 31)
column_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
input_df = pd.DataFrame([input_data], columns=column_names)

scaled_data = scaler.transform(input_df)

prediction = svc.predict(scaled_data)

if prediction == 1:
    print("This person has diabetes")
else:
    print("This person does not have diabetes")
```
